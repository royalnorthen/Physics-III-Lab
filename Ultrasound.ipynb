```python
import numpy as np
import matplotlib.pyplot as plt

time = np.array([2.6, 4.2, 5.8, 7.6, 8.8, 10.2, 11.7, 13.6, 13.8])
ultrasound = np.array([66.79, 57.78, 50.30, 42.75, 34.52, 26.24, 19.16, 65.47, 64.11])
ruler = np.array([64.0, 56.0, 48.0, 40.0, 32.0, 24.0, 16.0, 62.5, 59.5])

uncert_time = np.full_like(time, 0.1) 
uncert_ult = np.full_like(ultrasound, 0.04)
uncert_rul = np.full_like(ruler, 0.5)

# plotting
LabelFont = {'family':'serif','color':'black','size':20}
TitleFont = {'family':'serif','color':'black','size':30}

fig1 = plt.figure(figsize=(16,8))
ax1 = fig1.add_subplot(111)
ax1.errorbar(time, ultrasound, xerr=uncert_time, yerr=uncert_ult, fmt="o", label="Ultrasound")
ax1.errorbar(time, ruler, xerr=uncert_time, yerr=uncert_rul, fmt="o", label="Ruler")


ax1.xaxis.set_tick_params(labelsize=20)
ax1.yaxis.set_tick_params(labelsize=20)
ax1.set_xlabel("Time of Ultrasound Recording (s)", fontdict=LabelFont)
ax1.set_ylabel("Depth (mm)", fontdict=LabelFont)
ax1.set_title("Depth Measured by Ultrasound and by Ruler", fontdict=TitleFont)
```
