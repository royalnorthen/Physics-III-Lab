```python
import numpy as np
import matplotlib.pyplot as plt

attenuation_constants = np.array([2.0341, 4.0025, 5.8860, 7.6071, 9.2246])
std_dev = np.array([0.0072, 0.0121, 0.0167, 0.0209, 0.0235])

dye = np.arange(1, 6)
dye_uncert = np.arange(0.05, 0.30, 0.05)

# plotting
LabelFont = {'family':'serif','color':'black','size':20}
TitleFont = {'family':'serif','color':'black','size':30}

fig1 = plt.figure(figsize=(16,8))
ax1 = fig1.add_subplot(111)
ax1.scatter(dye, attenuation_constants, s=15, c='k')
ax1.errorbar(dye, attenuation_constants, xerr=dye_uncert, yerr=std_dev, fmt="o")


ax1.xaxis.set_tick_params(labelsize=20)
ax1.yaxis.set_tick_params(labelsize=20)
ax1.set_xlabel("Total Dye (mL)", fontdict=LabelFont)
ax1.set_ylabel("Attenuation Constant (m^-1)", fontdict=LabelFont)
ax1.set_title("Total Dye in Jar versus Attenuation Constant", fontdict=TitleFont)



```
