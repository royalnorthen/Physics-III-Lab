```python
import numpy as np
import matplotlib.pyplot as plt

# Experiment A
B = (0.001 * np.array([2.7, 3.1, 3.3, 3.6, 3.9, 4.1]))
r = (0.0279 + 0.01 * np.array([2.05, 3.05, 4.10, 5.25, 6.25, 7.25]))

uncert_B = np.full_like(B, 0.0001)
uncert_r = np.full_like(r, 0.0005)

slope, intercept = np.polyfit(B, r, 1)

print(slope)

# plotting
LabelFont = {'family':'serif','color':'black','size':20}
TitleFont = {'family':'serif','color':'black','size':30}

fig1 = plt.figure(figsize=(16,8))
ax1 = fig1.add_subplot(111)
ax1.errorbar(B, r, xerr=uncert_B, yerr=uncert_r, fmt="o")
ax1.plot(B, slope*B+intercept)  # fit line


ax1.xaxis.set_tick_params(labelsize=20)
ax1.yaxis.set_tick_params(labelsize=20)
ax1.set_xlabel("Magnetic Field (T)", fontdict=LabelFont)
ax1.set_ylabel("Moment Arm (m)", fontdict=LabelFont)
ax1.set_title("r versus B", fontdict=TitleFont)

```

```python
# Experiment B
inverseB = (0.001 * np.array([1.4, 2.0, 2.7, 3.4, 4.1, 4.8]))**-1  # Teslas ^ -1
period_squared = (0.05 * np.array([33.2, 26.2, 22.7, 20.8, 18.5, 17.4]))**2  # s ^ 2

uncert_B = inverseB / 14
uncert_T = np.full_like(period_squared, 0.2)  

slope, intercept = np.polyfit(inverseB, period_squared, 1)

print(slope)

# plotting
fig1 = plt.figure(figsize=(16,8))
ax1 = fig1.add_subplot(111)
ax1.errorbar(inverseB, period_squared, xerr=uncert_B, yerr=uncert_T, fmt="o")
ax1.plot(inverseB, slope*inverseB+intercept)  # fit line


ax1.xaxis.set_tick_params(labelsize=20)
ax1.yaxis.set_tick_params(labelsize=20)
ax1.set_xlabel("Magnetic Field^-1 (T^-1)", fontdict=LabelFont)
ax1.set_ylabel("Period^2 (s^2)", fontdict=LabelFont)
ax1.set_title("T^2 versus B^-1", fontdict=TitleFont)
```
